```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('../data/processed/dataset.csv')
print(df.head())
df.info()
df.describe()
```

```python
print(df.dtypes)
df.select_dtypes(include=["object"]).head()
```

```python
df["epoch_cal"].head()
```

```python
df["pha"].value_counts()
```

```python
# Drop ID columns
df.drop(columns=["id","spkid","full_name","pdes","name","orbit_id"], inplace=True)

# Convert categorical
cat_cols = ["prefix","neo","pha","class","equinox"]
df[cat_cols] = df[cat_cols].astype("category")

# Check missing values
print(df.isnull().sum())
```

```python
df["epoch_cal"] = pd.to_datetime(df["epoch_cal"].astype(int), format="%Y%m%d")
print(df["epoch_cal"].head())
```

```python
df["epoch_year"] = df["epoch_cal"].dt.year
df["epoch_month"] = df["epoch_cal"].dt.month
df["epoch_day"] = df["epoch_cal"].dt.day
df.drop(columns=["epoch_cal"], inplace=True)

```

```python
print(df["pha"].unique())
print(df["pha"].dtype)
print(df["pha"].isnull().sum())
```

```python
df["pha"].unique()

```

```python
df["pha"] = df["pha"].astype(str).str.strip().str.upper()


```

```python
df["pha"].value_counts(dropna=False)
```

```python
df["pha"] = df["pha"].map({"N":0, "Y":1})
df["pha"].value_counts()
```

```python
df.drop(columns=["prefix"], inplace=True)

```

```python
df.info()
```

```python
size_cols = ["diameter","albedo","diameter_sigma"]

for col in size_cols:
    df[size_cols] = df[size_cols].fillna(df[size_cols].median())
```

```python
df["neo"].fillna(df["neo"].mode()[0], inplace=True)
```

```python
df.drop(columns=["epoch","epoch_mjd"], inplace=True)
```

```python
df["neo"] = df["neo"].cat.codes
df["equinox"] = df["equinox"].cat.codes
```

```python
df.info()
```

```python

df.isnull().sum()
```

```python
cols = ["pha", "H", "moid", "moid_ld", "ma", "ad", "per", "per_y", "rms"]

for col in cols:
    df[col] = df[col].fillna(df[col].median())



```

```python
df.isnull().sum()
```

```python
sigma_cols = [col for col in df.columns if "sigma" in col]
df.drop(columns=sigma_cols, inplace=True)

```
